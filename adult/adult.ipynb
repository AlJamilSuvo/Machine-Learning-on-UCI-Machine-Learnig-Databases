```python
import pandas as pd
```

```python
column_names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']
```

```python
raw_dataset1=pd.read_csv('data/adult.data.csv', names=column_names, na_values='?', sep=', ')
raw_dataset2=pd.read_csv('data/adult.test.csv',names=column_names, na_values='?', sep=', ')
raw_dataset1=raw_dataset1.dropna()
raw_dataset2=raw_dataset2.dropna()
raw_dataset=pd.concat([raw_dataset1,raw_dataset2],ignore_index=True)
raw_dataset.tail()
```

```python
test_train_split_index=raw_dataset1.shape[0]
test_train_split_index
```

```python
series=pd.Series(raw_dataset['salary'])
series=series.map({'<=50K':0,'>50K':1,'<=50K.':0,'>50K.':1})
raw_dataset['salary']=series
raw_dataset['age']=pd.to_numeric(raw_dataset['age'])
raw_dataset.tail()
```

```python
dataset=pd.get_dummies(raw_dataset)
dataset
```

```python
dataset.describe().transpose()
```

```python
train_dataset=dataset[:test_train_split_index]
test_dataset=dataset[test_train_split_index:]
print(train_dataset.shape)
print(test_dataset.shape)
train_dataset.to_csv('data/processed/train.csv')
test_dataset.to_csv('data/processed/test.csv')
```

```python
train_dataset_pos=train_dataset[train_dataset['salary']==1]
train_dataset_neg=train_dataset[train_dataset['salary']==0]
print(train_dataset_pos.shape)
print(train_dataset_neg.shape)
```

```python
# train_dataset_neg_truncate=train_dataset_neg.sample(frac=train_dataset_pos.shape[0]/train_dataset_neg.shape[0],random_state=0)
# train_dataset_neg_truncate
```

```python
balaced_train_dataset=pd.concat([train_dataset_pos,train_dataset_neg, train_dataset_pos, train_dataset_pos])
balaced_train_dataset=balaced_train_dataset.sample(frac=1,random_state=0)
balaced_train_dataset
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers.experimental.preprocessing import Normalization
```

```python
train_features=balaced_train_dataset.copy()
test_features=test_dataset.copy()
train_label=train_features.pop('salary')
test_label=test_features.pop('salary')
train_features.shape
```

```python
train_label.describe()
```

```python
normalized=Normalization(input_shape=[104,])
```

```python
normalized.adapt(np.asarray(train_features))
```

```python
normalized.mean.numpy()
```

```python
print(normalized(np.asarray(train_features[:1])).numpy())
```

```python
model=Sequential()
model.add(normalized)
for i in range(3):
    model.add(Dense(12,activation='relu'))
    model.add(Dropout(rate=.5))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.predict(train_features[:10])
```

```python
model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(learning_rate=.0001),metrics=['accuracy'])
```

```python
history=model.fit(train_features,train_label,validation_split=.2, epochs=50)
```

```python
from plot_helper import plt_accuracy_learning_curve, plt_loss_learning_curve
```

```python
plt_accuracy_learning_curve(history)
plt_loss_learning_curve(history)
```

```python
predict_classes=model.predict_classes(test_features)
predict_classes
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_label,predict_classes))
```
